```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def longcall(ST,K,call):
  return np.maximum(ST-K,0) - call
def shortcall(ST,K,call):
  return -(np.maximum(ST-K,0) - call)
def longput(ST,K,put):
  return np.maximum(K-ST,0) - put
def shortput(ST,K,put):
  return -(np.maximum(K-ST,0) - put)
```

### Q1


```python
ST = np.arange(10000, 16000)
```

```python
SC13600 = shortcall(ST, 13600, 289)
SC14000 = shortcall(ST, 14000, 120)

plt.plot(ST, SC13600, 'r-', ST, SC14000, 'b-', [10000, 16000], [0, 0], 'k--')
```

```python
LP15000 = longput(ST, 15000, 1470)
LP14000 = longput(ST, 14000, 565)

plt.plot(ST, LP15000, 'r-', ST, LP14000, 'b-', [10000, 16000], [0, 0], 'k--')
```

```python
LC15000 = longcall(ST, 15000, 8.5)
SC13000 = shortcall(ST, 13000, 675)

LC15000 = longcall(ST, 15000, 8.5)
SC14000 = shortcall(ST, 14000, 120)


plt.plot(ST, LC15000+SC13000, 'r-', ST, LC15000+SC14000, 'b-', [10000, 16000], [0, 0], 'k--')
```

```python
price = [(13000 + i*100) for i in range(21)]
call = [675, 610 ,525, 448, 395, 345, 289, 238, 194, 153, 120, 92, 69, 52, 39.5, 28, 20.5, 16, 12, 9.8, 8.5]
put = [139, 163, 189, 221, 255, 291, 338, 388, 439, 505, 565, 640, 730, 800, 900, 995, 1090, 1180, 1290, 1380, 1470]
ST = np.arange(13000, 15000)
max_arbitrage = 0
price1 = 0
price2 = 0

for i in range(len(price)):
  for j in range(len(price)):
    SC1 = shortcall(ST, price[i], call[i])
    LP1 = longput(ST, price[i], put[i])
    SC2 = shortcall(ST, price[j], call[j])
    LP2 = longput(ST, price[j], put[j])
    if max(SC1+LP1) - max(SC2+LP2) > max_arbitrage:
      max_arbitrage = max(SC1+LP1) - max(SC2+LP2) 
      price1 = price[i]
      price2 = price[j]
print(f'buy: {price1}, sell: {price2}, max_arbitrage = {max_arbitrage}')
plt.plot(ST, shortcall(ST, price1, call[price.index(price1)])+longput(ST, price1, put[price.index(price1)]), 'r-', ST, shortcall(ST, price2, call[price.index(price2)])+longput(ST, price2, put[price.index(price2)]), 'b-', [13000, 15000], [0, 0], 'k--')
```
